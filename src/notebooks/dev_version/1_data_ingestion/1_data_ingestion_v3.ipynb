Remove full empty columns

```python
final.shape
```

```python
final = final.dropna(axis=1, how='all')
```

```python
final.shape
```

Remove unique value columns

```python
for i in final:
    if final[i].nunique() == 1:
        final = final.drop(columns=i)
```
